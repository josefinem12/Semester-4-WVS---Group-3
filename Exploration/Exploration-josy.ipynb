```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_justifiability_over_time(df, country_code, variable):
    """
    Plots the change in justifiability of an ethical issue over time for a specific country.
    
    Parameters:
        df (DataFrame): The WVS dataset.
        country_code (str): The country code (e.g., 'CHN', 'VEN', 'TUR', 'NLD').
        variable (str): The column name for the ethical issue (e.g., 'F114A', 'F118').
    """
    # Group by country and filter for the selected one
    grouped_df = df.groupby('COUNTRY_ALPHA')
    
    if country_code not in grouped_df.groups:
        print(f"No data available for {country_code}")
        return
    
    country_data = grouped_df.get_group(country_code)
    
    # Convert the selected variable to numeric, drop missing values
    country_data[variable] = pd.to_numeric(country_data[variable], errors='coerce')
    
    # Aggregate mean scores per year (S020)
    aggregated_country_data = country_data.groupby('S020')[variable].mean().reset_index()
    
    # Plot the trend
    plt.figure(figsize=(10, 6))
    plt.plot(aggregated_country_data['S020'], aggregated_country_data[variable], marker='o', linestyle='-')
    plt.title(f'Change in {variable} over time for {country_code}')
    plt.xlabel('Year (S020)')
    plt.ylabel('Mean Justifiability Score (1-10)')
    plt.grid(True)
    plt.show()

# Example usage:
# Assuming df is already loaded with the WVS dataset
plot_justifiability_over_time(df, 'CHN', 'F123')  # Homosexuality in China
plot_justifiability_over_time(df, 'VEN', 'F123')  # Abortion in Venezuela
plot_justifiability_over_time(df, 'TUR', 'F123')  # Bribery in Turkey
plot_justifiability_over_time(df, 'NLD', 'F123')  # Suicide in Netherlands
```
